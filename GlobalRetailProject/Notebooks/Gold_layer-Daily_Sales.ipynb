```python
spark.sql("USE globalretail_gold")
spark.sql("""
CREATE OR REPLACE TABLE gold_daily_sales AS
SELECT 
    transaction_date,
    ROUND(SUM(total_amount), 2) AS daily_total_sales
FROM 
    globalretail_silver.silver_orders
GROUP BY 
    transaction_date
""")
```

```python
display(spark.sql("select * from gold_daily_sales"))
```

```python
spark.sql("USE globalretail_gold")
spark.sql("""
CREATE OR REPLACE TABLE gold_category_sales AS
SELECT 
    p.category AS product_category,
    ROUND(SUM(o.total_amount), 2) AS category_total_sales
FROM 
    globalretail_silver.silver_orders o
JOIN 
    globalretail_silver.silver_products p ON o.product_id = p.product_id
GROUP BY 
    p.category
""")
```

```python
display(spark.sql("select * from gold_category_sales"))
```
