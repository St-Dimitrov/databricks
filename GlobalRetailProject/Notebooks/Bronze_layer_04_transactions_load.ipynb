```python
filePath = "dbfs:/FileStore/GlobalRetail/bronze/transactions/transaction_snappy.parquet"
df = spark.read.parquet(filePath)
df.show()
```

```python
from pyspark.sql.functions import to_timestamp, col
new_df = df.withColumn("transaction_date", to_timestamp(col("transaction_date")))
new_df.printSchema()
```

```python
display(new_df)
```

```python
from pyspark.sql.functions import current_timestamp
new_df = new_df.withColumn("ingestion_timestamp", current_timestamp())
display(new_df)
```

```python
spark.sql("use globalretail_bronze")
new_df.write.format("delta").mode("append").saveAsTable("bronze_transactions")
```

```python
spark.sql("use bronze_transactions")
```
